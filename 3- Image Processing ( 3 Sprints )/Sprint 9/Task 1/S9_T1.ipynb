### Edge detection using Sobel and Laplacian methods

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt  
%matplotlib inline
    
# Capture livestream video content from camera 0 
''' 
- cv2.VideoCapture(0):
   - Means first camera or webcam.
- cv2.VideoCapture(1):
   - Means second camera or webcam.
- cv2.VideoCapture("file name.mp4"):
   - Means video file
'''  
cap = cv2.VideoCapture(0)

while(1): 
  
    # Take each frame 
    _, frame = cap.read() 
      
    # Convert to RGB  
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) 
      
    # Calcution of Sobelx 
    # cv.Sobel(src, ddepth, xorder, yorder, dst, kernelsize, scale, delta, borderType)
    '''
    src : input image.
    ddepth : output image depth, in the case of 8-bit input images it will result in truncated derivatives.
    dx : order of the derivative x.
    dy : order of the derivative y.
    ksize : size of the extended Sobel kernel; it must be 1, 3, 5, or 7.
    scale : optional scale factor for the computed derivative values; by default, no scaling is applied (see getDerivKernels for details).
    delta : optional delta value that is added to the results prior to storing them in dst.
    borderType : pixel extrapolation method, see BorderTypes. BORDER_WRAP is not supported.
    '''
    sobelx = cv2.Sobel(rgb_frame, cv2.CV_64F, 1, 0, ksize = 5) 
      
    # Calculation of Sobely 
    # cv.Sobel(src, ddepth, xorder, yorder, dst, kernelsize, scale, delta, borderType)
    sobely = cv2.Sobel(rgb_frame, cv2.CV_64F, 0, 1, ksize = 5) 
      
    # Calculation of Laplacian 
    # cv.Laplacian(src, ddepth, dst, ksize, scale, delta, borderType)
    laplacian = cv2.Laplacian(rgb_frame,cv2.CV_64F) 
    
    cv2.imshow("Original", rgb_frame)
    cv2.imshow('Sobel-X',sobelx) 
    cv2.imshow('Sobel-Y',sobely) 
    cv2.imshow('Laplacian',laplacian) 
    k = cv2.waitKey(5) & 0xFF
    if k == 27: 
        break
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,20))
ax1.set_title('Original')
ax1.imshow(rgb_frame)
ax2.set_title('Sobel-X')
ax2.imshow(sobelx)
ax3.set_title('Sobel-Y ')
ax3.imshow(sobely)
ax4.set_title('Laplacian ')
ax4.imshow(laplacian)
            
cv2.destroyAllWindows() 
  
#release the frame 
cap.release() 
```

### Edge Detection using Sobel and Laplacian

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt  
%matplotlib inline

img = cv2.imread("Images/windows.jpg")

# Convert to RGB  
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

# Calcution of Sobelx 
# cv.Sobel(src, ddepth, xorder, yorder, dst, kernelsize, scale, delta, borderType)
sobelx = cv2.Sobel(rgb_img, cv2.CV_64F, 1, 0, ksize = 5) 

# Calculation of Sobely 
# cv.Sobel(src, ddepth, xorder, yorder, dst, kernelsize, scale, delta, borderType)
sobely = cv2.Sobel(rgb_img, cv2.CV_64F, 0, 1, ksize = 5) 

# Calculation of Laplacian 
# cv.Laplacian(src, ddepth, dst, ksize, scale, delta, borderType)
laplacian = cv2.Laplacian(rgb_img,cv2.CV_64F) 

cv2.imshow("Original", rgb_img)
cv2.imshow('Sobel-X',sobelx) 
cv2.imshow('Sobel-Y',sobely) 
cv2.imshow('Laplacian',laplacian) 

# De-allocate any associated memory usage   
if cv2.waitKey(0) & 0xff == 27:  
    cv2.destroyAllWindows()
        
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,20))
ax1.set_title('Original')
ax1.imshow(rgb_img)
ax2.set_title('Sobel-X')
ax2.imshow(sobelx)
ax3.set_title('Sobel-Y ')
ax3.imshow(sobely)
ax4.set_title('Laplacian ')
ax4.imshow(laplacian)
```

### Real-Time Edge Detection using OpenCV

```python
import cv2  
import numpy as np 
import matplotlib.pyplot as plt  
%matplotlib inline
    
# capture frames from a camera 
''' 
- cv2.VideoCapture(0):
   - Means first camera or webcam.
- cv2.VideoCapture(1):
   - Means second camera or webcam.
- cv2.VideoCapture("file name.mp4"):
   - Means video file
''' 
cap = cv2.VideoCapture(0) 
    
# loop runs if capturing has been initialized 
while(1): 
  
    # Capture frame-by-frame 
    # retval, frame = cv.VideoCapture.read(video)
    # Returns false if no frames has been grabbed
    ret, frame = cap.read() 
    
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) 
    
    # converting BGR to HSV 
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV) 
      
    # define range of red color in HSV 
    lower_red = np.array([30,150,50]) 
    upper_red = np.array([255,255,180]) 
      
    # create a red HSV colour boundary and  
    # threshold HSV image 
    # cv.inRange(src, lowerb, upperb, dst)
    # Detect an object based on the range of pixel values in the HSV colorspace
    mask = cv2.inRange(hsv, lower_red, upper_red) 
  
    # Bitwise-AND mask and original image 
    # # cv2.bitwise_and(source1, source2, destination, mask)
    res = cv2.bitwise_and(frame, frame, mask= mask) 
  
    # Display an original image 
    cv2.imshow('Original',rgb_frame) 
    
    # finds edges in the input image image and 
    # marks them in the output map edges 
    edges = cv2.Canny(rgb_frame, 100, 200) 
  
    # Display edges in a frame 
    cv2.imshow('Edges',edges) 
  
    # Wait for Esc key to stop 
    k = cv2.waitKey(5) & 0xFF
    if k == 27: 
        break

# Close the window 
cap.release() 
  
# De-allocate any associated memory usage 
cv2.destroyAllWindows() 

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,20))
ax1.set_title('Original')
ax1.imshow(rgb_frame)
ax2.set_title('Bitwise-AND')
ax2.imshow(res)
ax3.set_title('edges')
ax3.imshow(edges, cmap="gray")
```

### Line detection in python with OpenCV | Houghline method

  - The Hough Transform is a method that is used in image processing to detect any shape

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt  
%matplotlib inline
  
# Read image 
# cv2.imread(path, flag)
img = cv2.imread('Images/building.jpg') 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

  
# Convert the img to grayscale 
# cv2.cvtColor(src, code, dst, dstCn)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 

# Define our parameters for Canny
low_threshold = 50
high_threshold = 200

# Apply edge detection method on the image 
# cv.Canny(image, threshold1, threshold2, edges, apertureSize, L2gradient)
edges = cv2.Canny(gray, low_threshold, high_threshold, apertureSize = 3) 
  
# This returns an array of r and theta values 
# cv.HoughLines(image, rho, theta, threshold-votes, lines, srn, stn, min_theta, max_theta)
# Each line is represented by a 2 or 3 element vector (ρ,θ) or (ρ,θ,votes)

rho = 1 # Distance resolution
theta = np.pi/180 # Angle resolution
threshold = 80
min_line_length = 80
max_line_gap = 10

line_image = np.copy(img)

# cv.HoughLinesP(image, rho, theta, threshold, lines, minLineLength, maxLineGap)
lines = cv2.HoughLinesP(edges, rho, theta, threshold,
                        np.array([]), min_line_length,
                        max_line_gap) 

# Iterate over the output "lines" and draw lines on the image copy
for line in lines:
    for x1, y1, x2, y2 in line:
        #cv2.line(image, start_point, end_point, color, thickness)
        cv2.line(line_image,(x1,y1),(x2,y2),(255,0,0),3)        
        
# # The below for loop runs till r and theta values  
# # are in the range of the 2d array 
# for r,theta in lines[0]: 
      
#     # Stores the value of cos(theta) in a 
#     a = np.cos(theta) 
  
#     # Stores the value of sin(theta) in b 
#     b = np.sin(theta) 
      
#     # x0 stores the value rcos(theta) 
#     x0 = a*r 
      
#     # y0 stores the value rsin(theta) 
#     y0 = b*r 
      
#     # x1 stores the rounded off value of (rcos(theta)-1000sin(theta)) 
#     x1 = int(x0 + 1000*(-b)) 
      
#     # y1 stores the rounded off value of (rsin(theta)+1000cos(theta)) 
#     y1 = int(y0 + 1000*(a)) 
  
#     # x2 stores the rounded off value of (rcos(theta)+1000sin(theta)) 
#     x2 = int(x0 - 1000*(-b)) 
      
#     # y2 stores the rounded off value of (rsin(theta)-1000cos(theta)) 
#     y2 = int(y0 - 1000*(a)) 
      
#     # cv2.line draws a line in img from the point(x1,y1) to (x2,y2). 
#     # (0,0,255) denotes the colour of the line to be  
#     # drawn. In this case, it is red.  
#     # cv2.line(image, start_point, end_point, color, thickness)
#     cv2.line(img_copy,(x1,y1), (x2,y2), (0,0,255),2) 
      
# # All the changes made in the input image are finally 
# # written on a new image houghlines.jpg 
# cv2.imwrite('linesDetected.jpg', img) 


f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,20))
ax1.set_title('Original')
ax1.imshow(img)
ax2.set_title('EdgesDetected')
ax2.imshow(edges, cmap= 'gray')
ax3.set_title('linesDetected')
ax3.imshow(line_image)
```
