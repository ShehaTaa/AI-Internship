### 2. Extracting the RGB values of a pixel

```python
# import opencv 
import cv2

# v2.imread(path, flag)
img = cv2.imread('Images/road.jpg')

# Extracting RGB values.  
# Here we have randomly chosen a pixel 
# by passing in 100, 100 for height and width. 
(B, G, R) = img[100, 100] 
  
# Displaying the pixel values 
print("R = {}, G = {}, B = {}".format(R, G, B)) 
  
# We can also pass the channel to extract  
# the value for a specific channel 
value = img[100, 100, 0] 
print("B = {}".format(value)) 
```

### 4.  Resizing the Image

```python
# import opencv 
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline

# v2.imread(path, flag)
img = cv2.imread('Images/road.jpg')
# convert image from BGR format to RGB format
# cv2.cvtColor(src, code, dst, dstCn)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


# resize(src, dst, dsize, fx, fy, interpolation) 
resize = cv2.resize(img, (800, 800)) 

# Calculating the ratio 
h, w = img.shape[:2] 
ratio = 800 / w 
  
# Creating a tuple containing width and height 
dim = (800, int(h * ratio)) 
  
# Resizing the image 
resize_aspect = cv2.resize(img, dim) 

cv2.imshow("image", img)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,20))
ax1.set_title('Resizing image')
ax1.imshow(resize)
ax2.set_title('Resize Aspect')
ax2.imshow(resize_aspect)
```

### 5. Rotating the Image

```python
# import opencv 
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline

# v2.imread(path, flag)
img = cv2.imread('Images/road.jpg')
# convert image from BGR format to RGB format
# cv2.cvtColor(src, code, dst, dstCn)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Extracting the height and width of an image 
h, w = img.shape[:2] 

# Calculating the center of the image 
center = (w // 2, h // 2) 
  
# Generating a rotation matrix 
# cv2.getRotationMatrix2D(center, angle, scale)
'''
The function calculates the following matrix:

[α   β   (1−α)⋅center.x−β⋅center.y
-β  α   β⋅center.x+(1−α)⋅center.y]

where: α = scale⋅cos(angle), β = scale⋅sin(angle)
'''
matrix = cv2.getRotationMatrix2D(center, -45, 1.0)  
  
# Performing the affine transformation 
# cv2.warpAffine(src, Matrix, dsize, dst, flags, borderMode, borderValue)
rotated = cv2.warpAffine(img, matrix, (w, h)) 

'''
plt.imshow(X, cmap=None, norm=None, aspect=None,
           interpolation=None, alpha=None, vmin=None,
           vmax=None, origin=None, extent=None, shape=,
           filternorm=1, filterrad=4.0, imlim=, resample=None,
           url=None, \*, data=None, \*\*kwargs)
'''
plt.imshow(rotated)
```

### 6. Drawing a Rectangle

```python
# import opencv 
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline

# cv2.imread(path, flag)
img = cv2.imread('Images/road.jpg')
# convert image from BGR format to RGB format
# cv2.cvtColor(src, code, dst, dstCn)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# We are copying the original image,  
# as it is an in-place operation. 
output = img.copy() 
  
'''
Using the rectangle() function to create a rectangle. 
cv2.rectangle(Image, Top-left corner co-ordinates, 
              Bottom-right corner co-ordinates,
              Color (in BGR format), Line width)
''' 
rectangle = cv2.rectangle(output, (1500, 900),  
                          (600, 400), (255, 255, 0), 8) 

plt.imshow(rectangle)
```

### 7. Displaying text

```python
# import opencv 
import cv2
import matplotlib.pyplot as plt 
%matplotlib inline

# v2.imread(path, flag)
img = cv2.imread('Images/road.jpg')
# convert image from BGR format to RGB format
# cv2.cvtColor(src, code[, dst[, dstCn]])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Copying the original image 
output = img.copy() 
  
'''
Adding the text using putText() function 
cv.putText(Image, Text to be displayed, Bottom-left corner co-ordinates, from where the text should start,
           Font, Font size, Color (BGR format), Line width)

'''
text = cv2.putText(output, 'Welcome to the HELL', (500, 550),  
                   cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 255, 0), 8)

plt.imshow(text)
```

### 8. Reading an image in OpenCV

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
%matplotlib inline
  
# You can give path to the 
# image as first argument 
img = cv2.imread('Images/road.jpg') 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
 
plt.imshow(img)    
# will show the image in a window 
cv2.imshow('image', img)
# cv2.waitkey(delay)
k = cv2.waitKey(0) 
  
# wait for ESC key to exit 
# # The ASCII code corresponding to ESC is 27
if k == 27:  
    # simply destroys all the windows we created.
    cv2.destroyAllWindows() 
      
# wait for 's' key to save and exit 
elif k == ord('s'):  
    # cv2.imwrite(filename, image)
    cv2.imwrite('messigray.png',img) 
    cv2.destroyAllWindows() 
```

### 9. Addition of Image

```python
import cv2  
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
    
# path to input images are specified and   
# images are loaded with imread command  
img1 = cv2.imread('Images/img1.jpg')  
img2 = cv2.imread('Images/img2.jpg') 

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# cv2.addWeighted is applied over the 
# image inputs with applied parameters
# different weights are given to images in order to give a feeling of blending or transparency.
# cv2.addWeighted(img1, wt1, img2, wt2, gammaValue)
# output = α⋅img1 + β⋅img2 + γ
# α & β --> 0 : 1 
weightedSum = cv2.addWeighted(img1, 0.7, img2, 0.8, 0) 

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,20))
ax1.set_title('Image 1')
ax1.imshow(img1)
ax2.set_title('Image 2 ')
ax2.imshow(img2)
ax3.set_title("weighted Image")
ax3.imshow(weightedSum)


# the window showing output image 
# with the weighted sum  
cv2.imshow('Weighted Image', weightedSum) 
  
# De-allocate any associated memory usage 
# cv2.waitkey(delay)
if cv2.waitKey(0) & 0xff == 27:  
    cv2.destroyAllWindows() 
```

### 10. Subtraction of Image

```python
import cv2  
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
    
# path to input images are specified and   
# images are loaded with imread command  
img1 = cv2.imread('Images/imgstar1.jpg')  
img2 = cv2.imread('Images/imgstar2.jpg') 

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
  
# cv2.subtract is applied over the 
# image inputs with applied parameters 
# cv2.subtract(src1, src2)
sub = cv2.subtract(img1, img2) 

f, (ax1, ax2,ax3) = plt.subplots(1, 3, figsize=(20,20))
ax1.set_title('Image 1')
ax1.imshow(img1)
ax2.set_title('Image 2')
ax2.imshow(img2)
ax3.set_title('sub image')
ax3.imshow(sub)   
# the window showing output image 
# with the subtracted image  
cv2.imshow('Subtracted Image', sub) 
  
# De-allocate any associated memory usage   
if cv2.waitKey(0) & 0xff == 27:  
    cv2.destroyAllWindows()  
```

### 11. Bitwise AND operation on Image

   - Calculates the per-element bit-wise conjunction of two arrays or an array and a scalar.

```python
import cv2  
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
     
# path to input images are specified and   
# images are loaded with imread command  
img1 = cv2.imread('Images/bitwiseimg1.png')  
img2 = cv2.imread('Images/bitwiseimg2.png') 
  
# cv2.bitwise_and is applied over the 
# image inputs with applied parameters  
# cv2.bitwise_and(source1, source2, destination, mask)
dest_and = cv2.bitwise_and(img2, img1, mask = None) 
 
f, (ax1, ax2,ax3) = plt.subplots(1, 3, figsize=(20,20))
ax1.set_title('Image 1')
ax1.imshow(img1)
ax2.set_title('Image 2')
ax2.imshow(img2)
ax3.set_title('Bitwise And')
ax3.imshow(dest_and)       
# the window showing output image 
# with the Bitwise AND operation 
# on the input images 
cv2.imshow('Bitwise And', dest_and) 
   
# De-allocate any associated memory usage   
if cv2.waitKey(0) & 0xff == 27:  
    cv2.destroyAllWindows() 
```

### 12. Bitwise OR operation on Image

   - Calculates the per-element bit-wise disjunction of two arrays or an array and a scalar.

```python
import cv2  
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
     
# path to input images are specified and   
# images are loaded with imread command  
img1 = cv2.imread('Images/bitwiseimg1.png')  
img2 = cv2.imread('Images/bitwiseimg2.png') 
  
# cv2.bitwise_or is applied over the 
# image inputs with applied parameters  
# cv2.bitwise_or(source1, source2, destination, mask)
dest_or = cv2.bitwise_or(img2, img1, mask = None) 
 
f, (ax1, ax2,ax3) = plt.subplots(1, 3, figsize=(20,20))
ax1.set_title('Image 1')
ax1.imshow(img1)
ax2.set_title('Image 2')
ax2.imshow(img2)
ax3.set_title('OR image')
ax3.imshow(dest_or)  
# the window showing output image 
# with the Bitwise OR operation 
# on the input images 
cv2.imshow('Bitwise OR', dest_or) 
   
# De-allocate any associated memory usage   
if cv2.waitKey(0) & 0xff == 27:  
    cv2.destroyAllWindows()  
```

### 14. Bitwise NOT operation on Image

```python
import cv2  
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
     
# path to input images are specified and   
# images are loaded with imread command  
img1 = cv2.imread('Images/bitwiseimg1.png')  
img2 = cv2.imread('Images/bitwiseimg2.png') 
  
# cv2.bitwise_not is applied over the 
# image input with applied parameters  
# cv2.bitwise_not(source, destination, mask)
dest_not1 = cv2.bitwise_not(img1, mask = None) 
dest_not2 = cv2.bitwise_not(img2, mask = None) 
 
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,20))
ax1.set_title('Image 1')
ax1.imshow(img1)
ax2.set_title('Dest Not1')
ax2.imshow(dest_not1)
ax3.set_title('Image 2')
ax3.imshow(img2)
ax4.set_title('Dest Not2')
ax4.imshow(dest_not2)  

# the windows showing output image 
# with the Bitwise NOT operation 
# on the 1st and 2nd input image 
cv2.imshow('Bitwise NOT on image 1', dest_not1) 
cv2.imshow('Bitwise NOT on image 2', dest_not2) 
   
# De-allocate any associated memory usage   
if cv2.waitKey(0) & 0xff == 27:  
    cv2.destroyAllWindows() 
```
