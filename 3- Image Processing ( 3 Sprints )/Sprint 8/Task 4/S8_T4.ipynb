### 1. Play a video using OpenCV

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
   
# Create a VideoCapture object and read from input file 
# cv2.VideoCapture(index or file_name)
cap = cv2.VideoCapture('Data/music.mp4') 
   
# Check if camera opened successfully 
# retval = cv.VideoCapture.isOpened() const
# Returns true if video capturing has been initialized already.
if (cap.isOpened()== False):  
    print("Error opening video  file") 
    
# Read until video is completed 
while(cap.isOpened()): 
      
    # Capture frame-by-frame 
    # retval, frame = cv.VideoCapture.read(video)
    # Returns false if no frames has been grabbed
    ret, frame = cap.read() 
    
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
    if ret == True: 
        # Display the resulting frame 
        #cv2.imshow('Frame', frame) 
        plt.imshow(frame)
   
    # Press Q on keyboard to  exit 
    if cv2.waitKey(25) & 0xFF == ord('q'): 
        break
   
  # Break the loop 
    else:  
        break
        
# When everything done, release  
# the video capture object 
# Closes video file or capturing device
cap.release() 
   
# Closes all the frames 
cv2.destroyAllWindows() 
```

### 2. Extract images from video

```python
import cv2 
import os 
import matplotlib.pyplot as plt  
%matplotlib inline

# Read the video from specified path 
# cv2.VideoCapture(index or file_name)
cam = cv2.VideoCapture('Data/music.mp4') 
  
try:      
    # creating a folder named data 
    if not os.path.exists('Data'): 
        os.makedirs('Data') 
        
# if not created then raise error 
except OSError: 
    print ('Error: Creating directory of Data') 
    
# frame 
currentframe = 0
  
while(True): 
      
    # reading from frame 
    ret,frame = cam.read() 
  
    if ret: 
        # if video is still left continue creating images 
        name = './Data/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name) 
  
        # writing the extracted images 
        # retval = cv.imwrite(filename, img, params) --> Saves an image to a specified file.
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        cv2.imwrite(name, frame) 
  
        # increasing counter so that it will 
        # show how many frames are created 
        currentframe += 1
    else: 
        break
        
# Release all space and windows once done 
cam.release() 
cv2.destroyAllWindows() 

f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,20))
ax1.set_title('Frame 1')
ax1.imshow(cv2.imread('Data/frame0.jpg'))
ax2.set_title('Frame 2')
ax2.imshow(cv2.imread('Data/frame1.jpg'))
ax3.set_title('Frame 3')
ax3.imshow(cv2.imread('Data/frame2.jpg'))
ax4.set_title('Frame 4')
ax4.imshow(cv2.imread('Data/frame3.jpg'))
```

### 3. Face Detection using Python and OpenCV

```python
import cv2
import matplotlib.pyplot as plt  
%matplotlib inline

# Load the cascade
# retval = cv.CascadeClassifier.load(filename)
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# To capture video from webcam. 
# cv2.VideoCapture(index or file_name)
cap = cv2.VideoCapture(0)

while True:
    # Read the frame
    # retval, image = cv.VideoCapture.read(image)
    _, img = cap.read()

    # Detect the faces
    # cv2.CascadeClassifier.detectMultiScale(image, scaleFactor, minNeighbors, flags, minSize, maxSize)
    '''
     scaleFactor : Parameter specifying how much the image size is reduced at each image scale.
     minNeighbors : Parameter specifying how many neighbors each candidate rectangle should have to retain it.
     minSize : Minimum possible object size. Objects smaller than that are ignored.
     maxSize : Maximum possible object size. Objects larger than that are ignored.
    '''
    face_detected = face_cascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=20)

    # Draw the rectangle around each face
    for (x, y, w, h) in face_detected:
        # cv2.rectangle(image, start_point, end_point, color, thickness)
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 4)

    # Display
    cv2.imshow('img', img)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)

    # Stop if escape key is pressed
    k = cv2.waitKey(30) & 0xff
    if k==27:
        break
        
# Release the VideoCapture object
cap.release()
```
